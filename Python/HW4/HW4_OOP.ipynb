```python
class Set:
    def __init__(self, inputList = []):
        for i in inputList:
            if inputList.count(i) > 1:
                inputList.remove(i)
        self.inputList = inputList
        
    def add(self, elem):
        if elem not in self:
            self.inputList.append(elem)
            
    def discard(self, elem):
        if elem in self:
            self.inputList.remove(elem)
    
    def clear(self):
        self.inputList.clear()
    
    def __len__(self):
        return len(self.inputList)
    
    def __str__(self):
        return "{{{}}}".format(str(self.inputList).strip('[').strip(']'))
    
    def __contains__(self, elem):
        for i in self.inputList:
            if i == elem:
                return True
        return False
        
    def __le__(self, other):
        for i in self.inputList:
            if i not in other.inputList:
                return False
        return True
    
    def __ge__(self, other):
        for i in other.inputList:
            if i not in self.inputList:
                return False
        return True
    
    def __or__(self, other):
        temp = Set([])
        for i in self.inputList:
            temp.add(i)
        for j in other.inputList:
            temp.add(j)
        return temp
    
    def __and__(self, other):
        temp2 = Set([])
        for i in self.inputList:
            if i in other:
                temp2.add(i)
        return temp2
    
    def __sub__(self, other):
        temp3 = Set([])
        for i in self.inputList:
            if i not in other:
                temp3.add(i)
        return temp3
    
    def __iand__(self, other):
        temp = self & other
        self.clear()
        for i in temp.inputList:
            self.add(i)
        return self
    
    def __ior__(self, other):
        for i in other.inputList:
            self.add(i)
        return self
    
    def __isub__(self, other):
        for i in self.inputList:
            if i in other:
                self.discard(i)
        return self
    
a = Set([1, 2, 3, 4])
b = Set([1, 2, 3, 4])

print(a)
print(b)
print()

a.discard(4)
b.discard(1)

print(a)
print(b)
print()

print(len(a))
print(1 in a)
print(1 in b)
print()

print(a | b)
print(a & b)
print(a - b)
print()

print(a <= b)
print(a <= a | b)
print(a >= b)
print(a >= a & b)
print()

b.clear()
print(b)
print()

a = Set([1, 2, 3])
b = Set([3, 4])
address_a = id(a)
a |= b
print(a)
print(address_a == id(a))
print()

a = Set([1, 2, 3])
b = Set([3, 4])
address_a = id(a)
a &= b
print(a)
print(address_a == id(a))
print()

a = Set([1, 2, 3])
b = Set([3, 4])
address_a = id(a)
a -= b
print(a)
print(address_a == id(a))
print()
```
