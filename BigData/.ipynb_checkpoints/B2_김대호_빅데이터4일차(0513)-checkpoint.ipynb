## 실습 3번

```python
import pandas as pd
import numpy as np

pop_seoul = pd.read_csv('2. 데이터 Handling/서울인구통계.txt', sep='\t', engine='python', thousands=',', encoding='utf-8')
pop_seoul.head()
```

```python
pop1 = pop_seoul[['자치구', '인구', '인구.3', '인구.6', '65세이상고령자']]
pop_seoul = pop1.drop([0, 1])
pop_seoul.head()
```

```python
pop_seoul.rename(columns={pop_seoul.columns[0]: '구별',
                         pop_seoul.columns[1]: '인구수',
                         pop_seoul.columns[2]: '한국인',
                         pop_seoul.columns[3]: '외국인',
                         pop_seoul.columns[4]: '고령자',}, inplace=True)
pop_seoul.head()
```

```python
cctv_seoul = pd.read_excel('2. 데이터 Handling/서울시CCTV.xlsx')
cctv_seoul.head()
```

```python
cctv_seoul.columns
```

```python
cctv_seoul.rename(columns={cctv_seoul.columns[0] : '구별'}, inplace=True)
cctv_seoul.rename(columns={cctv_seoul.columns[1] : 'cctv수'}, inplace=True)
cctv_seoul.head()
```

```python
cctv_seoul.sort_values(by='cctv수', ascending=True).head(10)
```

```python
pop_seoul['구별'].unique()
```

```python
cctv_seoul.isnull().sum()
```

```python
cctv_seoul.loc[cctv_seoul.isnull()['2011년 이전']]
```

```python
cctv_seoul["최근증가율"]=(
    cctv_seoul["2018년"]+
    cctv_seoul["2017년"]+
    cctv_seoul["2016년"])/(cctv_seoul["cctv수"]
                          -(cctv_seoul["2018년"]+
                            cctv_seoul["2017년"]+
                            cctv_seoul["2016년"]))*100
cctv_seoul.sort_values(by='최근증가율', ascending=False).head(5)
```

```python
pop_seoul.drop([2], inplace=True)
pop_seoul
```

```python
pop_seoul.dtypes
```

```python
pop_seoul=pop_seoul.astype({"인구수":"int","한국인":"int","외국인":"int","고령자":"int"})
```

```python
pop_seoul['외국인 비율'] = pop_seoul['외국인'] / pop_seoul['인구수']*100
pop_seoul['고령자 비율'] = pop_seoul['고령자'] / pop_seoul['인구수']*100
pop_seoul.head()
```

```python
pop_seoul.sort_values(by="외국인 비율", ascending=False).head()
```

```python
pop_seoul.sort_values(by="고령자 비율", ascending=False).head()
```

```python
cctv_seoul["구별"]=["강남구","강동구","강북구","강서구","관악구","광진구","구로구","금천구","노원구","도봉구",
"동대문구","동작구","마포구","서대문구","서초구","성동구","성북구","송파구","양천구","영등포구","용산구","은평구",
"종로구","중구","중랑구"]
```

```python
result = pd.merge(cctv_seoul, pop_seoul, on="구별")
result
```

```python
del result['2011년 이전']
del result['2012년']
del result['2013년']
del result['2014년']
del result['2015년']
del result['2016년']
del result['2017년']
del result['2018년']
result.head()
```
